```python
import pandas as pd
import numpy as np
import pickle as pk
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MinMaxScaler
```

Import and Cleaning Data

```python
df = pd.read_csv('portfolios.csv')
```

```python
df.describe()
```

```python
df_check = df.isnull().values.any()
print(df_check)
```

```python
df.dropna(axis=0,how='any',inplace=True)
df.drop(columns='index',inplace=True)
```

```python
df
```

```python
df['SMA42'] = MinMaxScaler().fit_transform(df[['SMA42']])
df['SMA150'] = MinMaxScaler().fit_transform(df[['SMA150']])
df['RSI'] = MinMaxScaler().fit_transform(df[['RSI']])
df['VAR'] = MinMaxScaler().fit_transform(df[['VAR']])
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df.describe()
```

```python
df
```

Creating Knn Model

```python
knn = NearestNeighbors(n_neighbors=10)
knn.fit(df.drop(columns='stocks'))
```

Generating Recomendations

```python
neighbors = knn.kneighbors(df.drop(columns='stocks'),n_neighbors=20)
```

```python
df.drop(columns='stocks')
```

```python
type(df['stocks'][0])
```

```python
def recomend(index, stocks, neighbors, num):
    whallet = stocks[index]
    whallet = whallet[1:len(whallet)-1].replace("'","").split(',')
    to_recomend = {}
    for x in neighbors[index]:
        neighbor = stocks[x]
        neighbor = neighbor[1:len(neighbor)-1].replace("'","").split(',')
        for y in neighbor:
            if y not in whallet:
                elm = to_recomend.get(y)
                if elm:
                    to_recomend.update({y:elm+1})
                else:
                    to_recomend.update({y:1})

    sorted_arr = np.array(list({k:v for k,v in sorted(to_recomend.items(),key=lambda item:item[1])}.items()))
    return sorted_arr[len(sorted_arr)-num:]

```
